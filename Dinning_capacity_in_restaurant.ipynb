```python

import math

def calculate_queue_metrics(lambda_rate, mu_rate, c_servers, scenario_name):
    print(f"--- WORKINGS FOR {scenario_name.upper()} ---")

    # Step 1: Calculate Total Capacity
    total_capacity = c_servers * mu_rate
    print(f"Step 1: Total System Capacity (c * mu) = {c_servers} * {mu_rate:.2f} = {total_capacity:.2f} parties/hour")

    # Step 2: Calculate Utilization (rho)
    rho = lambda_rate / total_capacity
    print(f"Step 2: Utilization (rho = lambda / (c * mu)) = {lambda_rate} / {total_capacity:.2f} = {rho:.2f} ({rho*100:.1f}%)")

    if rho >= 1.0:
        print("Error: System is unstable (Utilization >= 100%). The line will grow infinitely!")
        return

    # Step 3: Calculate P0 (Probability that the system is entirely empty)
    # Formula: 1 / [ Sum_{n=0}^{c-1} ((lambda/mu)^n / n!) + ((lambda/mu)^c / (c! * (1 - rho))) ]
    traffic_intensity = lambda_rate / mu_rate

    sum_terms = 0
    for n in range(c_servers):
        sum_terms += (traffic_intensity ** n) / math.factorial(n)

    overflow_term = (traffic_intensity ** c_servers) / (math.factorial(c_servers) * (1 - rho))
    p0 = 1 / (sum_terms + overflow_term)
    print(f"Step 3: Probability that the system is completely empty (P0) = {p0:.4f}")

    # Step 4: Calculate Erlang's C Formula (Probability that an arrival has to wait in line)
    # Formula: P_Q = [ (lambda/mu)^c / (c! * (1 - rho)) ] * P0
    p_wait = overflow_term * p0
    print(f"Step 4: Probability that a new arrival has to wait in line (P_wait) = {p_wait:.4f} ({p_wait*100:.1f}%)")

    # Step 5: Calculate Average Wait Time in Queue (W_q)
    # Formula: W_q = P_wait / (c * mu - lambda)
    w_q_hours = p_wait / (total_capacity - lambda_rate)
    w_q_minutes = w_q_hours * 60
    print(f"Step 5: Average time spent waiting in the queue (W_q) = {w_q_minutes:.1f} minutes")

    # Step 6: Calculate Total Time in the Restaurant (W = W_q + 1/mu)
    service_time_minutes = (1 / mu_rate) * 60
    total_time_minutes = w_q_minutes + service_time_minutes
    print(f"Step 6: Total time at restaurant (Wait time + {service_time_minutes:.1f} min dining time) = {total_time_minutes:.1f} minutes\n")

# Run the two scenarios from the slides
# Average dining time is 45 minutes -> mu = 60 / 45 = 1.3333 tables cleared per hour
mu = 4 / 3
tables = 4

# Scenario A: Calm / Manageable
calculate_queue_metrics(lambda_rate=3, mu_rate=mu, c_servers=tables, scenario_name="Scenario A (Calm Night)")

# Scenario B: Chaos / Busy Friday
calculate_queue_metrics(lambda_rate=5, mu_rate=mu, c_servers=tables, scenario_name="Scenario B (Busy Friday Night)")
```
