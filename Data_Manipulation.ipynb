```python
#add total col as sum of individuals and family_members
homelessness["total"] = homelessness["individuals"] + homelessness["family_members"]
```

```python
#Add p_homelessness col as proportion of total homeless population to the state population
homelessness["p_homelessness"] = homelessness["total"] / homelessness["state_pop"]
```

```python
print(homelessness)
```

```python
#create indiv_per_10k as homeless individuals per 10k state pop
homelessness["indiv_per_10k"] = 10000 * homelessness["individuals"] / homelessness["state_pop"]
```

```python
#Subset rows for indiv_per_10k greater than 20
high_homelessness = homelessness[homelessness["indiv_per_10k"] > 20]
```

```python
#sort high_homelessness by descending indiv_per_10k
high_homelessness_srt = high_homelessness.sort_values("indiv_per_10k", ascending=False)
```

```python
#From the high_homelessness_srt, select the state and indiv_per_10k cols
result = high_homelessness_srt[["state", "indiv_per_10k"]]
```

```python
import pandas as pd

# This is a reliable mirror of the Walmart sales subset
url = "https://raw.githubusercontent.com/OlajideFemi/Carbon-Footprint/refs/heads/main/train.csv"

# Load the data
sales = pd.read_csv(url)

# The subset you were using had lowercase names;
# Let's rename them so your existing code works perfectly!
sales.columns = [col.lower() for col in sales.columns]

# Print the head to confirm it matches your expected output
print(sales.head())

# Now your calculation will work
print(f"\nMean Weekly Sales: {sales['weekly_sales'].mean()}")
print(f"Median Weekly Sales: {sales['weekly_sales'].median()}")
```

```python
# A custom IQR function
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

# Update to print IQR of temperature_c, fuel_price_usd_per_l, & unemployment
print(sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg(iqr))
```

```python
#Update the aggregation functions called by .agg(): include iqr and "median" in that order.
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25), column.median()

print(sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr, "median"]))
```

```python
# Create a custom IQR function
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
def median(column):
    return column.median()

# Update to print IQR and median of temperature_c, fuel_price_usd_per_l, & unemployment
print(sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr,median]))
```
