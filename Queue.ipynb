```python

import pandas as pd
import numpy as np

# Set a random seed so the results are identical every time you run it
np.random.seed(42)

# 1. Row count
num_rows = 100

# 2. Generate features
user_ids = np.arange(1, num_rows + 1)
ages = np.random.randint(18, 71, size=num_rows)
income = np.random.normal(loc=50000, scale=15000, size=num_rows).round(2)
tiers = ['Silver', 'Gold', 'Platinum']
membership = np.random.choice(tiers, size=num_rows, p=[0.5, 0.3, 0.2])

# 3. Create DataFrame
synthetic_data = pd.DataFrame({
    'UserID': user_ids,
    'Age': ages,
    'Annual_Income': income,
    'Membership_Tier': membership
})

# 4. Output
print(synthetic_data.head())
synthetic_data.to_csv('synthetic_user_data.csv', index=False)
```
