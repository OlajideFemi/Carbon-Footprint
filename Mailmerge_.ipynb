```python
Â #Advanced Template with Dynamic Attachments
function advancedMailMerge() {
    var sheet = SpreadsheetApp.getActiveSheet();
      var data = sheet.getDataRange().getValues();
        var headers = data[0];

            var template = {
                subject: "Important document for {{firstName}}",
                    body: `Dear {{firstName}},

                    Your personalized document is attached.

                    Additional details:
                    - Date: {{currentDate}}
                    - Reference: {{referenceNumber}}

                    Best regards,
                    Your Team`
                      };

                          for (var i = 1; i < data.length; i++) {
                              var row = data[i];
                                  var email = row[headers.indexOf('Email')];
                                      var firstName = row[headers.indexOf('FirstName')];
                                          var attachmentUrl = row[headers.indexOf('AttachmentURL')];
                                              var reference = row[headers.indexOf('Reference')];

                                                      if (!email) continue;

                                                              // Replace placeholders
                                                                  var subject = template.subject.replace('{{firstName}}', firstName);
                                                                      var body = template.body
                                                                            .replace(/{{firstName}}/g, firstName)
                                                                                  .replace('{{currentDate}}', new Date().toLocaleDateString())
                                                                                        .replace('{{referenceNumber}}', reference);

                                                                                                // Send with attachment
                                                                                                    var file = getFileFromUrl(attachmentUrl);

                                                                                                            GmailApp.sendEmail(email, subject, body, {
                                                                                                                  attachments: [file.getAs(MimeType.PDF)],
                                                                                                                        name: 'Your Company Name'
                                                                                                                            });

                                                                                                                                    // Mark as sent and add timestamp
                                                                                                                                        sheet.getRange(i + 1, headers.indexOf('Sent') + 1).setValue(new Date());
                                                                                                                                            Utilities.sleep(1200); // Avoid rate limits
                                                                                                                                              }
                                                                                                                                              }
}
```

```python
function sendPersonalizedEmails() {
    var sheet = SpreadsheetApp.getActiveSheet();
      var dataRange = sheet.getDataRange();
        var data = dataRange.getValues();
          var headers = data[0];

              // Find column indexes
                var emailIndex = headers.indexOf('Email');
                  var firstNameIndex = headers.indexOf('FirstName');
                    var attachmentIndex = headers.indexOf('AttachmentURL');
                      var sentIndex = headers.indexOf('Sent');

                          // Start from row 1 (skip headers)
                            for (var i = 1; i < data.length; i++) {
                                var row = data[i];
                                    var email = row[emailIndex];
                                        var firstName = row[firstNameIndex];
                                            var attachmentUrl = row[attachmentIndex];
                                                var sent = row[sentIndex];

                                                        // Skip if already sent or email is empty
                                                            if (sent || !email) continue;

                                                                    try {
                                                                          // Send email
                                                                                sendEmail(email, firstName, attachmentUrl);

                                                                                            // Mark as sent
                                                                                                  sheet.getRange(i + 1, sentIndex + 1).setValue('YES');

                                                                                                              // Optional: Add delay to avoid rate limits
                                                                                                                    Utilities.sleep(1000);

                                                                                                                              } catch (error) {
                                                                                                                                    Logger.log('Error sending to ' + email + ': ' + error.toString());
                                                                                                                                        }
                                                                                                                                          }
                                                                                                                                          }

                                                                                                                                          function sendEmail(email, firstName, attachmentUrl) {
                                                                                                                                            var subject = `Hello ${firstName}! Important Document`;
                                                                                                                                              var body = `
                                                                                                                                              Dear ${firstName},

                                                                                                                                              Please find your personalized document attached.

                                                                                                                                              Best regards,
                                                                                                                                              Your Team
                                                                                                                                                `;

                                                                                                                                                    var file = getFileFromUrl(attachmentUrl);

                                                                                                                                                        MailApp.sendEmail({
                                                                                                                                                            to: email,
                                                                                                                                                                subject: subject,
                                                                                                                                                                    body: body,
                                                                                                                                                                        attachments: [file.getAs(MimeType.PDF)]
                                                                                                                                                                          });
                                                                                                                                                                          }

                                                                                                                                                                          function getFileFromUrl(url) {
                                                                                                                                                                            try {
                                                                                                                                                                                var fileId = extractFileIdFromUrl(url);
                                                                                                                                                                                    return DriveApp.getFileById(fileId);
                                                                                                                                                                                      } catch (error) {
                                                                                                                                                                                          throw new Error('Could not access file: ' + url);
                                                                                                                                                                                            }
                                                                                                                                                                                            }

                                                                                                                                                                                            function extractFileIdFromUrl(url) {
                                                                                                                                                                                              var match = url.match(/[-\w]{25,}/);
                                                                                                                                                                                                return match ? match[0] : null;
                                                                                                                                                                                                }
}
```
