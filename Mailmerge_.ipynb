```python
function advancedMailMerge() {
    var sheet = SpreadsheetApp.getActiveSheet();
      var data = sheet.getDataRange().getValues();
        var headers = data[0];

            var template = {
                subject: "Important document for {{firstName}}",
                    body: `Dear {{firstName}},

                    Your personalized document is attached.

                    Additional details:
                    - Date: {{currentDate}}
                    - Reference: {{referenceNumber}}

                    Best regards,
                    Your Team`
                      };

                          for (var i = 1; i < data.length; i++) {
                              var row = data[i];
                                  var email = row[headers.indexOf('Email')];
                                      var firstName = row[headers.indexOf('FirstName')];
                                          var attachmentUrl = row[headers.indexOf('AttachmentURL')];
                                              var reference = row[headers.indexOf('Reference')];

                                                      if (!email) continue;

                                                              // Replace placeholders
                                                                  var subject = template.subject.replace('{{firstName}}', firstName);
                                                                      var body = template.body
                                                                            .replace(/{{firstName}}/g, firstName)
                                                                                  .replace('{{currentDate}}', new Date().toLocaleDateString())
                                                                                        .replace('{{referenceNumber}}', reference);

                                                                                                // Send with attachment
                                                                                                    var file = getFileFromUrl(attachmentUrl);

                                                                                                            GmailApp.sendEmail(email, subject, body, {
                                                                                                                  attachments: [file.getAs(MimeType.PDF)],
                                                                                                                        name: 'Your Company Name'
                                                                                                                            });

                                                                                                                                    // Mark as sent and add timestamp
                                                                                                                                        sheet.getRange(i + 1, headers.indexOf('Sent') + 1).setValue(new Date());
                                                                                                                                            Utilities.sleep(1200); // Avoid rate limits
                                                                                                                                              }
                                                                                                                                              }
}
```
