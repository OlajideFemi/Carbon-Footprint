Seaports: The Logistical Backbone of Global Trade

A port is a maritime facility comprising one or more wharves or loading areas, where ships load and discharge cargo and passengers. Although usually situated on a sea coast or estuary, ports can also be found far inland, such as Hamburg, Manchester and Duluth; these access the sea via rivers or canals. Because of their roles as ports of entry for immigrants as well as soldiers in wartime, many port cities have experienced dramatic multi-ethnic and multicultural changes throughout their histories.

How could EU industrial action impact UK supply chain operations?




EU industrial action can significantly impact UK supply chain operations by disrupting the intermodal transfer process, where goods move between maritime transport and land-based logistics like trucks and trains. Because seaports function as critical gateways, any slowdown in the EU is likely to delay vessel arrivals at UK docks and berths, interrupting the 24/7 management coordinated by port authorities.

The specific impacts on UK operations include:

• Cargo Handling Inefficiency: Industrial action that reduces the availability of longshoremen or specialised equipment operators leads to immediate increases in waiting times and severe congestion at the wharf.

• Storage and Capacity Constraints: UK warehouses and storage yards are designed for the temporary storage of cargo. If industrial action in the EU halts the flow of goods, these areas—even those utilising RFID tracking technology—can quickly reach capacity.

• Queue Spillback: When "Internal Transport Systems" fail to move goods out of the port efficiently, it leads to queue spillback. This doesn't just stall the supply chain; it creates urban mobility issues in surrounding areas, affecting the "Port-City Relationship" and the quality of life for residents and the port's workforce.

• Regional Competitiveness: Since over 90% of international trade is conducted via maritime transport, inefficient operations at these hubs directly threaten the competitiveness of the affected region.

• Logistical Bottlenecks: The sources suggest that mitigating these impacts requires capacity mapping for specialised terminals and utilising flexible internal transport systems, such as railways or autonomous vehicles, to bypass truck-related congestion.

```python
def simulate_industrial_action(
    G, impacted, cap_reduction=0.6, delay_days=5, reroute=True
):
    """
    impacted: list of EU countries affected by industrial action
    cap_reduction: 0.6 means capacity becomes 40% of baseline (i.e., 60% reduced)
    delay_days: add delay proxy to impacted edges
    reroute: attempt to reroute overflow to alternative nodes
    """
    H = G.copy()

    # Add runtime attributes
    for u, v, data in H.edges(data=True):
        data["current_capacity"] = data["capacity"]
        data["current_shipments"] = data["baseline_shipments"]
        data["delay"] = 0.0

    # Apply shock on edges touching impacted nodes
    for u, v, data in H.edges(data=True):
        if u in impacted or v in impacted:
            data["current_capacity"] = data["capacity"] * (1 - cap_reduction)
            data["delay"] += delay_days

    # Compute overflow (demand - capacity)
    overflow = []
    for u, v, data in H.edges(data=True):
        demand = data["baseline_shipments"]
        cap = data["current_capacity"]
        if demand > cap:
            overflow_amt = demand - cap
            data["current_shipments"] = cap
            overflow.append((u, v, overflow_amt))
        else:
            data["current_shipments"] = demand

    # Reroute overflow by sending to alternative destination via a substitute node (very simple)
    if reroute and overflow:
        # substitutes: non-impacted EU + UK (exclude impacted)
        substitutes = [n for n in H.nodes() if n not in impacted]

        for u, v, amt in overflow:
            # Try reroute: u -> s -> v where s not impacted
            # choose s that minimizes u->s + s->v distance proxy (requires centroid dict in outer scope)
            best_s = None
            best_cost = 1e18
            for s in substitutes:
                if s == u or s == v:
                    continue
                if not (H.has_edge(u, s) and H.has_edge(s, v)):
                    continue
                # distance proxy:
                lon1, lat1 = centroids[u]
                lon2, lat2 = centroids[s]
                lon3, lat3 = centroids[v]
                cost = haversine_km(lon1, lat1, lon2, lat2) + haversine_km(lon2, lat2, lon3, lat3)
                if cost < best_cost:
                    best_cost = cost
                    best_s = s

            if best_s is not None:
                # push into u->s and s->v with capacity check
                for a, b in [(u, best_s), (best_s, v)]:
                    ed = H[a][b]
                    free = max(ed["current_capacity"] - ed["current_shipments"], 0)
                    moved = min(free, amt)
                    ed["current_shipments"] += moved
                    ed["delay"] += delay_days * 0.5  # reroute adds some delay
                    amt -= moved
                    if amt <= 0:
                        break
            # remaining amt is "lost"

    # UK impact metrics
    inbound_baseline = sum(H[u][UK]["baseline_shipments"] for u in H.predecessors(UK) if H.has_edge(u, UK))
    inbound_current  = sum(H[u][UK]["current_shipments"] for u in H.predecessors(UK) if H.has_edge(u, UK))
    lost_inbound = inbound_baseline - inbound_current

    # delay proxy: weighted avg delay on inbound UK edges
    inbound_edges = [(u, UK) for u in H.predecessors(UK) if H.has_edge(u, UK)]
    total_in = sum(H[u][UK]["current_shipments"] for u, _ in inbound_edges) + 1e-9
    avg_delay = sum(H[u][UK]["delay"] * H[u][UK]["current_shipments"] for u, _ in inbound_edges) / total_in

    return {
        "inbound_baseline_shipments": inbound_baseline,
        "inbound_current_shipments": inbound_current,
        "lost_inbound_shipments": lost_inbound,
        "avg_inbound_delay_days": avg_delay,
        "graph": H
    }

# Example scenario
impacted = ["France", "Belgium", "Netherlands", "Spain"]
result = simulate_industrial_action(G, impacted, cap_reduction=0.7, delay_days=7, reroute=True)
result
```

```python
# ============================
# EU-27 + UK Synthetic Trade Network (Bidirectional) + Industrial Action Shock
# Uses your shipping_data as the "realistic shipment attribute" sampler
# Inputs in this workspace:
#   /mnt/data/ne_110m_admin_0_countries.geojson
#   /mnt/data/shipping_data.csv.zip   (or /mnt/data/shipping_data.csv)
# ============================

import os
import numpy as np
import pandas as pd
import geopandas as gpd
import networkx as nx
from math import radians, sin, cos, asin, sqrt

# ----------------------------
# 0) Config
# ----------------------------
RNG_SEED = 42
np.random.seed(RNG_SEED)

EU27 = [
    "Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus",
    "Czech Republic", "Denmark", "Estonia", "Finland", "France", "Germany", "Greece",
    "Hungary", "Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg",
    "Malta", "Netherlands", "Poland", "Portugal", "Romania",
    "Slovakia", "Slovenia", "Spain", "Sweden"
]
UK = "United Kingdom"
COUNTRIES = EU27 + [UK]

GEOJSON_PATH = "ne_110m_admin_0_countries.geojson"
CSV_PATH = "shipping_data.csv"
ZIP_PATH = "shipping_data.csv.zip"

# Name differences between your list and Natural Earth (varies by file version)
ALIASES = {
    "Czech Republic": "Czechia",
}

# Some tiny countries can be missing at 110m (e.g., Malta). Provide fallback lon/lat.
FALLBACK_LONLAT = {
    "Malta": (14.3754, 35.9375),
}

# Synthetic shipment count (increase for smoother network)
N_SYNTH = 50000

# ----------------------------
# 1) Utilities
# ----------------------------
def haversine_km(lon1, lat1, lon2, lat2) -> float:
    lon1, lat1, lon2, lat2 = map(radians, [float(lon1), float(lat1), float(lon2), float(lat2)])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    return 2 * asin(sqrt(a)) * 6371.0

def load_shipping_df() -> pd.DataFrame:
    # Read CSV if present; otherwise read from ZIP (single-file zip supported by pandas)
    if os.path.exists(CSV_PATH):
        df = pd.read_csv(CSV_PATH)
    elif os.path.exists(ZIP_PATH):
        df = pd.read_csv(ZIP_PATH)
    else:
        raise FileNotFoundError(f"Could not find {CSV_PATH} or {ZIP_PATH}")

    # Basic cleaning
    # Keep key numeric columns; drop rows missing shipment core attributes
    required = ["name", "price ($)", "weight (kg)", "length (m)", "width (m)", "height (m)", "shipment date"]
    for c in required:
        if c not in df.columns:
            raise ValueError(f"Missing expected column: {c}. Found columns: {list(df.columns)}")

    df["shipment date"] = pd.to_datetime(df["shipment date"], errors="coerce")
    df = df.dropna(subset=["name", "price ($)", "weight (kg)"]).copy()
    df["shipment date"] = df["shipment date"].fillna(pd.Timestamp("2023-03-01"))
    return df

def build_country_centroids(geojson_path: str,
                            countries: list[str],
                            aliases: dict[str, str],
                            fallback_lonlat: dict[str, tuple[float, float]]) -> dict[str, tuple[float, float]]:
    world_all = gpd.read_file(geojson_path)

    # Resolve names for filtering polygons
    resolved = [aliases.get(c, c) for c in countries]
    world = world_all[world_all["ADMIN"].isin(resolved)].copy()

    # Compute centroids in projected CRS to avoid geographic centroid warning
    world_proj = world.to_crs(3857)
    cent_proj = world_proj.geometry.centroid
    cent_wgs84 = gpd.GeoSeries(cent_proj, crs=3857).to_crs(4326)

    # centroids keyed by ADMIN names in file
    centroids_admin = {admin: (pt.x, pt.y) for admin, pt in zip(world["ADMIN"], cent_wgs84)}

    # Build centroids keyed by YOUR country list (with alias + fallback)
    centroids_by_your_names = {}
    missing_no_fallback = []
    for c in countries:
        admin_name = aliases.get(c, c)
        if admin_name in centroids_admin:
            centroids_by_your_names[c] = centroids_admin[admin_name]
        elif c in fallback_lonlat:
            centroids_by_your_names[c] = fallback_lonlat[c]
        else:
            missing_no_fallback.append(c)

    if missing_no_fallback:
        raise KeyError(f"Missing centroid(s) with no fallback: {missing_no_fallback}")

    return centroids_by_your_names

# ----------------------------
# 2) Gravity model for OD sampling
# ----------------------------
def gravity_prob(orig: str, dest: str,
                 centroids: dict[str, tuple[float, float]],
                 beta: float = 1.2,
                 eps: float = 1e-6) -> float:
    lon1, lat1 = centroids[orig]
    lon2, lat2 = centroids[dest]
    d = haversine_km(lon1, lat1, lon2, lat2) + eps
    return 1.0 / (d ** beta)

def generate_synthetic_shipments(df_shipping: pd.DataFrame,
                                 countries: list[str],
                                 centroids: dict[str, tuple[float, float]],
                                 n: int,
                                 beta: float = 1.2) -> pd.DataFrame:
    # Build OD probability space
    pairs, probs = [], []
    for o in countries:
        for d in countries:
            if o == d:
                continue
            p = gravity_prob(o, d, centroids=centroids, beta=beta)
            pairs.append((o, d))
            probs.append(p)

    probs = np.array(probs, dtype=float)
    probs = probs / probs.sum()

    # Sample OD routes
    chosen_idx = np.random.choice(len(pairs), size=n, p=probs)
    routes = [pairs[i] for i in chosen_idx]

    # Sample shipment attributes (realistic distributions from your data)
    syn = df_shipping.sample(n=n, replace=True, random_state=RNG_SEED).reset_index(drop=True)
    syn["origin_country"] = [r[0] for r in routes]
    syn["dest_country"] = [r[1] for r in routes]

    # Resample dates into a 30-day window for a simple time axis
    syn["shipment date"] = syn["shipment date"] + pd.to_timedelta(np.random.randint(0, 30, size=n), unit="D")
    return syn

def aggregate_flows(syn: pd.DataFrame) -> pd.DataFrame:
    flows = (
        syn.groupby(["origin_country", "dest_country"], as_index=False)
           .agg(
                shipments=("name", "count"),
                total_weight_kg=("weight (kg)", "sum"),
                total_value_usd=("price ($)", "sum"),
            )
    )
    return flows

# ----------------------------
# 3) Build bidirectional directed graph
# ----------------------------
def build_graph_from_flows(flows: pd.DataFrame, countries: list[str], eu27: list[str], uk: str) -> nx.DiGraph:
    G = nx.DiGraph()
    for c in countries:
        G.add_node(c, is_eu=(c in eu27), is_uk=(c == uk))

    # Add edges with baseline + capacity
    for _, r in flows.iterrows():
        o, d = r["origin_country"], r["dest_country"]
        baseline_shipments = float(r["shipments"])
        capacity = baseline_shipments * 1.2  # simple headroom; tune as needed
        G.add_edge(
            o, d,
            baseline_shipments=baseline_shipments,
            baseline_weight=float(r["total_weight_kg"]),
            baseline_value=float(r["total_value_usd"]),
            capacity=float(capacity),
        )
    return G

# ----------------------------
# 4) Industrial action simulation
# ----------------------------
def simulate_industrial_action(
    G: nx.DiGraph,
    centroids: dict[str, tuple[float, float]],
    impacted: list[str],
    cap_reduction: float = 0.7,   # 0.7 => capacity reduced by 70% (keeps 30%)
    delay_days: float = 7.0,
    reroute: bool = True,
) -> dict:
    """
    Applies a shock to edges touching impacted nodes: capacity reduced + delays increased.
    Then caps flows to capacity and (optionally) reroutes overflow via a substitute node.

    Returns UK impact metrics + shocked graph.
    """
    H = G.copy()

    # init runtime attributes
    for u, v, data in H.edges(data=True):
        data["current_capacity"] = data["capacity"]
        data["current_shipments"] = data["baseline_shipments"]
        data["delay"] = 0.0

    impacted_set = set(impacted)

    # apply shock
    for u, v, data in H.edges(data=True):
        if (u in impacted_set) or (v in impacted_set):
            data["current_capacity"] = data["capacity"] * (1.0 - cap_reduction)
            data["delay"] += delay_days

    # cap demand to capacity; collect overflow
    overflow = []
    for u, v, data in H.edges(data=True):
        demand = data["baseline_shipments"]
        cap = data["current_capacity"]
        if demand > cap:
            overflow_amt = demand - cap
            data["current_shipments"] = cap
            overflow.append((u, v, overflow_amt))
        else:
            data["current_shipments"] = demand

    # reroute overflow via one intermediate substitute node s (u->s->v)
    if reroute and overflow:
        substitutes = [n for n in H.nodes() if n not in impacted_set]

        for u, v, amt in overflow:
            # Find best s that minimizes distance u->s + s->v, and edges exist
            best_s, best_cost = None, float("inf")
            for s in substitutes:
                if s in (u, v):
                    continue
                if not (H.has_edge(u, s) and H.has_edge(s, v)):
                    continue
                lon_u, lat_u = centroids[u]
                lon_s, lat_s = centroids[s]
                lon_v, lat_v = centroids[v]
                cost = haversine_km(lon_u, lat_u, lon_s, lat_s) + haversine_km(lon_s, lat_s, lon_v, lat_v)
                if cost < best_cost:
                    best_cost = cost
                    best_s = s

            if best_s is None:
                continue  # can't reroute; lost flow remains implicit

            # Push flow through u->s and s->v subject to remaining capacity
            for a, b in [(u, best_s), (best_s, v)]:
                ed = H[a][b]
                free = max(ed["current_capacity"] - ed["current_shipments"], 0.0)
                moved = min(free, amt)
                ed["current_shipments"] += moved
                ed["delay"] += delay_days * 0.5  # reroute adds extra delay
                amt -= moved
                if amt <= 0:
                    break

    # UK impact metrics (inbound)
    if UK not in H.nodes():
        raise KeyError("UK node not found in graph.")

    inbound_edges = [(u, UK) for u in H.predecessors(UK) if H.has_edge(u, UK)]
    inbound_baseline = sum(H[u][UK]["baseline_shipments"] for u, _ in inbound_edges) if inbound_edges else 0.0
    inbound_current = sum(H[u][UK]["current_shipments"] for u, _ in inbound_edges) if inbound_edges else 0.0
    lost_inbound = inbound_baseline - inbound_current

    total_in = inbound_current + 1e-9
    avg_delay = (sum(H[u][UK]["delay"] * H[u][UK]["current_shipments"] for u, _ in inbound_edges) / total_in) if inbound_edges else 0.0

    return {
        "impacted": impacted,
        "cap_reduction": cap_reduction,
        "delay_days": delay_days,
        "inbound_baseline_shipments": inbound_baseline,
        "inbound_current_shipments": inbound_current,
        "lost_inbound_shipments": lost_inbound,
        "avg_inbound_delay_days": avg_delay,
        "graph": H,
    }

# ----------------------------
# 5) Run end-to-end
# ----------------------------
# Build centroids for EU-27 + UK (with alias + fallback)
centroids_by_name = build_country_centroids(
    GEOJSON_PATH, COUNTRIES, aliases=ALIASES, fallback_lonlat=FALLBACK_LONLAT
)
print("Centroids ready. Example:", "Malta", centroids_by_name["Malta"], "| Czech Republic", centroids_by_name["Czech Republic"])

# Load shipping data
df_ship = load_shipping_df()
print("Shipping rows (clean):", len(df_ship))

# Generate synthetic shipments (EU27+UK OD routes with realistic attributes)
syn = generate_synthetic_shipments(df_ship, COUNTRIES, centroids_by_name, n=N_SYNTH, beta=1.2)
print("Synthetic shipments:", len(syn))
print(syn[["name","origin_country","dest_country","weight (kg)","price ($)","shipment date"]].head())

# Aggregate to flows
flows = aggregate_flows(syn)
print("OD flow rows:", len(flows))
print(flows.sort_values("shipments", ascending=False).head(10))

# Build bidirectional directed graph
G = build_graph_from_flows(flows, COUNTRIES, EU27, UK)
print("Graph:", G.number_of_nodes(), "nodes,", G.number_of_edges(), "edges")

# ----------------------------
# 6) Example industrial action scenario
# ----------------------------
impacted = ["France", "Belgium", "Netherlands", "Spain", "Germany"]  # edit as needed
result = simulate_industrial_action(
    G,
    centroids=centroids_by_name,
    impacted=impacted,
    cap_reduction=0.7,   # 70% capacity cut on edges touching impacted nodes
    delay_days=7.0,
    reroute=True,
)

print("\n=== UK Impact Metrics ===")
print("Impacted:", result["impacted"])
print("UK inbound baseline shipments:", round(result["inbound_baseline_shipments"], 2))
print("UK inbound current shipments :", round(result["inbound_current_shipments"], 2))
print("UK lost inbound shipments    :", round(result["lost_inbound_shipments"], 2))
print("UK avg inbound delay (days)  :", round(result["avg_inbound_delay_days"], 2))

# Optional: quick look at top reduced inbound edges to UK
H = result["graph"]
inbound_edges = [(u, UK) for u in H.predecessors(UK) if H.has_edge(u, UK)]
rows = []
for u, v in inbound_edges:
    rows.append({
        "from": u,
        "baseline": H[u][v]["baseline_shipments"],
        "current": H[u][v]["current_shipments"],
        "lost": H[u][v]["baseline_shipments"] - H[u][v]["current_shipments"],
        "delay": H[u][v]["delay"],
        "impacted_edge": (u in set(impacted)) or (v in set(impacted))
    })
inbound_df = pd.DataFrame(rows).sort_values("lost", ascending=False)
print("\nTop UK inbound losses (by sender country):")
print(inbound_df.head(10))
```

Percolation Theory

The Percolation Transition

When you remove nodes from a network, there is a critical point ($p_c$) where the network suddenly "breaks" into small, isolated islands.

Subcritical ($p < p_c$): The network is a series of isolated ports. No "flow" is possible across the Channel.

Critical ($p = p_c$): The "Tipping Point." One more failure causes the system to go from "struggling" to "collapsed."

Supercritical ( p_c" data-index-in-node="15" style="font-family: "Google Sans Text", sans-serif !important; line-height: 1.15 !important; margin-top: 0px !important;">$p > p_c$): A "Giant Component" exists. Even if one port is down, you can still get goods from the EU to the UK via alternative routes



```python
import networkx as nx
import matplotlib.pyplot as plt

def simulate_percolation():
    # 1. Setup Graph with Capacity and Load
    G = nx.Graph()
    port_data = {
        "Dover": {"load": 8000, "capacity": 9200},
        "Calais": {"load": 8000, "capacity": 9000},
        "Felixstowe": {"load": 7000, "capacity": 7800},
        "Rotterdam": {"load": 12000, "capacity": 13500},
        "Dunkirk": {"load": 2000, "capacity": 2500},
        "Antwerp": {"load": 11000, "capacity": 12000},
        "Zeebrugge": {"load": 4000, "capacity": 4800},
        "Hull": {"load": 3000, "capacity": 3600}
    }
    for p, d in port_data.items(): G.add_node(p, **d)

    # Define Bidirectional Connectivity
    G.add_edges_from([("Dover", "Calais"), ("Dover", "Dunkirk"), ("Calais", "Dunkirk"),
                      ("Felixstowe", "Rotterdam"), ("Felixstowe", "Antwerp"),
                      ("Rotterdam", "Antwerp"), ("Hull", "Zeebrugge"),
                      ("Hull", "Rotterdam"), ("Zeebrugge", "Antwerp")])

    gcc_history = [len(max(nx.connected_components(G), key=len))]
    failed = []
    queue = ["Calais"]

    while queue:
        node = queue.pop(0)
        if node not in G: continue

        load_to_move = G.nodes[node]['load']
        failed.append(node)
        G.remove_node(node)

        # Calculate new GCC Size
        if len(G) > 0:
            gcc_history.append(len(max(nx.connected_components(G), key=len)))
            # Redistribute load
            for n in G.nodes:
                G.nodes[n]['load'] += load_to_move / len(G.nodes)
                if G.nodes[n]['load'] > G.nodes[n]['capacity'] and n not in queue:
                    queue.append(n)
        else:
            gcc_history.append(0)

    return gcc_history, failed

# Execute and Plot
gcc_history, failure_order = simulate_percolation()
plt.plot(gcc_history, marker='o', color='red')
plt.title("Percolation: Destruction of the Giant Component")
plt.xlabel("Failure Step"); plt.ylabel("Nodes in Giant Component")
plt.savefig('percolation_plot.png')
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def simulate_percolation_cascade(initial_node):
    # 1. Setup the Graph
    # We use a slightly larger set of ports to see the percolation better
    G = nx.Graph()

    # Capacity = Load * (1 + alpha). Alpha = 0.15 for high vulnerability
    port_data = {
        "Dover": {"load": 8000, "capacity": 9200},
        "Calais": {"load": 8000, "capacity": 9000},
        "Felixstowe": {"load": 7000, "capacity": 7800},
        "Rotterdam": {"load": 12000, "capacity": 13500},
        "Dunkirk": {"load": 2000, "capacity": 2500},
        "Antwerp": {"load": 11000, "capacity": 12000},
        "Zeebrugge": {"load": 4000, "capacity": 4800},
        "Hull": {"load": 3000, "capacity": 3600}
    }

    for port, data in port_data.items():
        G.add_node(port, **data)

    # Define connections (Trade Routes)
    edges = [
        ("Dover", "Calais"), ("Dover", "Dunkirk"), ("Calais", "Dunkirk"),
        ("Felixstowe", "Rotterdam"), ("Felixstowe", "Antwerp"),
        ("Rotterdam", "Antwerp"), ("Hull", "Zeebrugge"),
        ("Hull", "Rotterdam"), ("Zeebrugge", "Antwerp")
    ]
    G.add_edges_from(edges)

    # Track metrics
    gcc_sizes = []
    failed_nodes = []

    # Helper to get GCC size
    def get_gcc_size(graph):
        if len(graph) == 0: return 0
        components = sorted(nx.connected_components(graph), key=len, reverse=True)
        return len(components[0])

    # Step 0: Initial State
    gcc_sizes.append(get_gcc_size(G))

    # Step 1: Initial Failure
    nodes_to_fail = [initial_node]

    while nodes_to_fail:
        current_node = nodes_to_fail.pop(0)
        if current_node not in G: continue

        orphaned_load = G.nodes[current_node]['load']
        failed_nodes.append(current_node)
        G.remove_node(current_node)

        # Calculate GCC after node removal
        current_gcc = get_gcc_size(G)
        gcc_sizes.append(current_gcc)

        # Redistribute load to surviving nodes
        remaining_nodes = list(G.nodes)
        if remaining_nodes:
            load_per_node = orphaned_load / len(remaining_nodes)
            for n in remaining_nodes:
                G.nodes[n]['load'] += load_per_node
                # Check for new failure
                if G.nodes[n]['load'] > G.nodes[n]['capacity']:
                    if n not in nodes_to_fail:
                        nodes_to_fail.append(n)
        else:
            break

    return gcc_sizes, failed_nodes

# Run Simulation
gcc_history, failure_order = simulate_percolation_cascade("Calais")

# Plotting the Percolation Graph
plt.figure(figsize=(10, 6))
steps = list(range(len(gcc_history)))
plt.plot(steps, gcc_history, marker='o', linestyle='-', color='red', linewidth=2, markersize=8)

# Labeling
plt.title("Percolation of Logistics Collapse: Giant Component Size vs. Failure Steps", fontsize=14)
plt.xlabel("Failure Step (0=Stable, 1=Calais Failure, 2+=Cascades)", fontsize=12)
plt.ylabel("Size of Giant Connected Component (Nodes)", fontsize=12)
plt.xticks(steps)
plt.grid(True, which='both', linestyle='--', alpha=0.5)

# Annotate nodes as they fail
for i, node in enumerate(["None"] + failure_order):
    if i < len(steps):
        plt.annotate(node, (steps[i], gcc_history[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('percolation_logistics_cascade.png')
print(f"Failure Order: {failure_order}")
print(f"GCC Sizes: {gcc_history}")
```

```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import time
import os
from pathlib import Path

class PortWatchAnalyzer:
    """Production-ready PortWatch data analyzer for disruption modelling"""

    def __init__(self, cache_dir='portwatch_cache'):
        self.base_url = "https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/Daily_Trade_Data/FeatureServer/0/query"
        self.cache_dir = Path(cache_dir)
        self.cache_dir.mkdir(exist_ok=True)

        # Known port IDs (you'll need to update these after discovery)
        self.port_ids = {
            # UK ports
            'Dover': 'PORT826',      # Verify these!
            'Folkestone': 'PORT827',
            'Portsmouth': 'PORT828',
            'Poole': 'PORT829',
            'Plymouth': 'PORT830',

            # French ports
            'Calais': 'PORT132',
            'Dunkerque': 'PORT133',
            'Le Havre': 'PORT134',
            'Cherbourg': 'PORT135',

            # Benelux
            'Rotterdam': 'PORT480',
            'Zeebrugge': 'PORT581',
            'Antwerp': 'PORT582',

            # Irish
            'Dublin': 'PORT230',
            'Rosslare': 'PORT231'
        }

    def date_to_timestamp(self, date_str):
        """Convert date string to ArcGIS timestamp"""
        dt = datetime.strptime(date_str, '%Y-%m-%d')
        return int(dt.timestamp() * 1000)

    def timestamp_to_date(self, ts):
        """Convert ArcGIS timestamp to datetime"""
        return datetime.fromtimestamp(ts / 1000)

    def fetch_port_data(self, port_id, start_date, end_date, use_cache=True):
        """
        Fetch daily data for a specific port with date range

        Parameters:
        - port_id: Port identifier
        - start_date: YYYY-MM-DD
        - end_date: YYYY-MM-DD
        - use_cache: Save/load from CSV cache
        """

        # Create cache filename
        cache_file = self.cache_dir / f"{port_id}_{start_date}_{end_date}.csv"

        # Check cache
        if use_cache and cache_file.exists():
            print(f"Loading cached data for {port_id}")
            return pd.read_csv(cache_file, parse_dates=['date'])

        # Convert dates to timestamps
        start_ts = self.date_to_timestamp(start_date)
        end_ts = self.date_to_timestamp(end_date)

        # Build where clause with date filter
        # Note: You'll need the actual date field name - commonly 'date', 'report_date', or 'timestamp'
        where_clause = f"portid='{port_id}' AND date >= {start_ts} AND date <= {end_ts}"

        params = {
            "where": where_clause,
            "outFields": "*",
            "maxRecordCountFactor": "5",  # Get up to 5000 records
            "outSR": "4326",
            "f": "json",
            "orderByFields": "date ASC"  # Get chronological order
        }

        print(f"Fetching {port_id} from {start_date} to {end_date}...")

        try:
            response = requests.get(self.base_url, params=params, timeout=30)
            response.raise_for_status()
            data = response.json()

            if 'error' in data:
                print(f"API Error for {port_id}: {data['error'].get('message', 'Unknown')}")
                return None

            features = data.get('features', [])
            if not features:
                print(f"No data for {port_id} in this period")
                return pd.DataFrame()

            # Convert to DataFrame
            records = []
            for f in features:
                record = f['attributes']
                # Convert timestamp to datetime if date field exists
                if 'date' in record:
                    record['date'] = self.timestamp_to_date(record['date'])
                records.append(record)

            df = pd.DataFrame(records)

            # Cache the results
            if use_cache and not df.empty:
                df.to_csv(cache_file, index=False)
                print(f"Cached to {cache_file}")

            return df

        except requests.exceptions.RequestException as e:
            print(f"Request failed for {port_id}: {e}")
            return None

    def analyze_fy24_25(self):
        """Pull and analyze full FY24/25 data for all key ports"""

        # Define financial year quarters
        fy24_q1 = ('2024-04-01', '2024-06-30')
        fy24_q2 = ('2024-07-01', '2024-09-30')
        fy24_q3 = ('2024-10-01', '2024-12-31')
        fy24_q4 = ('2025-01-01', '2025-03-31')

        fy25_q1 = ('2025-04-01', '2025-06-30')
        fy25_q2 = ('2025-07-01', '2025-09-30')
        fy25_q3 = ('2025-10-01', '2025-12-31')
        fy25_q4 = ('2026-01-01', '2026-03-31')

        quarters = {
            'FY24 Q1': fy24_q1,
            'FY24 Q2': fy24_q2,
            'FY24 Q3': fy24_q3,
            'FY24 Q4': fy24_q4,
            'FY25 Q1': fy25_q1,
            'FY25 Q2': fy25_q2,
            'FY25 Q3': fy25_q3,
            'FY25 Q4': fy25_q4
        }

        all_data = []

        # Fetch data for each port
        for port_name, port_id in self.port_ids.items():
            print(f"\n{'='*60}")
            print(f"Processing {port_name} ({port_id})")
            print('='*60)

            # Get full FY24/25 in one go if possible
            df = self.fetch_port_data(
                port_id,
                '2024-04-01',
                '2026-03-31',
                use_cache=True
            )

            if df is not None and not df.empty:
                df['port_name'] = port_name
                df['port_id'] = port_id
                all_data.append(df)

            time.sleep(1)  # Be polite to the API

        # Combine all data
        if all_data:
            combined_df = pd.concat(all_data, ignore_index=True)

            # Ensure date is datetime
            if 'date' in combined_df.columns:
                combined_df['date'] = pd.to_datetime(combined_df['date'])

                # Add time-based features
                combined_df['year'] = combined_df['date'].dt.year
                combined_df['month'] = combined_df['date'].dt.month
                combined_df['quarter'] = combined_df['date'].dt.quarter
                combined_df['day_of_week'] = combined_df['date'].dt.dayofweek
                combined_df['weekend'] = combined_df['day_of_week'].isin([5, 6])

                # Financial year
                combined_df['financial_year'] = combined_df['date'].apply(
                    lambda x: f"FY{str(x.year)[2:]}" if x.month >= 4 else f"FY{str(x.year-1)[2:]}"
                )

            # Save master file
            combined_df.to_csv('portwatch_fy24_25_master.csv', index=False)
            print(f"\nSaved master file with {len(combined_df)} records")

            return combined_df

        return pd.DataFrame()

# Run the analysis
analyzer = PortWatchAnalyzer()
fy24_25_data = analyzer.analyze_fy24_25()

if not fy24_25_data.empty:
    print("\n" + "="*80)
    print("FY24/25 DATA SUMMARY")
    print("="*80)

    # Summary by port
    summary = fy24_25_data.groupby('port_name').agg({
        'date': ['min', 'max', 'count'],
        # Add other metrics based on actual column names
    }).round(2)

    print("\nData coverage by port:")
    print(summary)

    # Daily traffic patterns
    if 'day_of_week' in fy24_25_data.columns:
        daily_pattern = fy24_25_data.groupby(['port_name', 'day_of_week']).size().unstack()
        print("\nWeekly patterns (records by day of week):")
        print(daily_pattern)

    # Save quarterly breakdowns
    fy24_25_data['year_quarter'] = fy24_25_data['date'].dt.to_period('Q')
    quarterly = fy24_25_data.groupby(['port_name', 'year_quarter']).size()
    quarterly.to_csv('quarterly_breakdown.csv')
    print("\nQuarterly breakdown saved to 'quarterly_breakdown.csv'")
```

```python
import requests
import pandas as pd
from datetime import datetime

def fetch_port_daily_data(service_name="Daily_Trade_Data_WLD", port_id=None, start_date="2024-04-01", end_date="2026-03-31"):
    """
    Fetch daily port data from PortWatch API

    Parameters:
    - service_name: One of the discovered services (Daily_Trade_Data_WLD, Daily_Ports_Data, etc.)
    - port_id: Optional specific port ID (if None, returns all ports)
    - start_date: Start date in YYYY-MM-DD
    - end_date: End date in YYYY-MM-DD
    """

    base_url = f"https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/{service_name}/FeatureServer/0/query"

    # Convert dates to ArcGIS timestamp format (milliseconds)
    def date_to_timestamp(date_str):
        dt = datetime.strptime(date_str, '%Y-%m-%d')
        return int(dt.timestamp() * 1000)

    start_ts = date_to_timestamp(start_date)
    end_ts = date_to_timestamp(end_date)

    # Build where clause
    where_parts = [f"date >= {start_ts}", f"date <= {end_ts}"]
    if port_id:
        where_parts.append(f"portid='{port_id}'")

    where_clause = " AND ".join(where_parts)

    params = {
        "where": where_clause,
        "outFields": "*",
        "returnGeometry": "false",
        "maxRecordCountFactor": "5",  # Get up to 5000 records
        "resultRecordCount": "5000",
        "orderByFields": "date ASC",
        "f": "json"
    }

    print(f"Fetching {service_name} data from {start_date} to {end_date}...")
    if port_id:
        print(f"Port ID: {port_id}")

    response = requests.get(base_url, params=params)

    if response.status_code != 200:
        print(f"Error: {response.status_code}")
        return pd.DataFrame()

    data = response.json()

    if 'error' in data:
        print(f"API Error: {data['error'].get('message')}")
        return pd.DataFrame()

    features = data.get('features', [])
    if not features:
        print("No data returned")
        return pd.DataFrame()

    # Convert to DataFrame
    records = []
    for f in features:
        attrs = f['attributes']
        # Convert timestamp to datetime if date field exists
        if 'date' in attrs and attrs['date']:
            attrs['date'] = datetime.fromtimestamp(attrs['date'] / 1000)
        records.append(attrs)

    df = pd.DataFrame(records)
    print(f"Retrieved {len(df)} records")

    return df

# 1. First, let's find available port IDs from the ports database
print("Getting port IDs from PortWatch_ports_database...")
ports_df = fetch_port_daily_data(
    service_name="PortWatch_ports_database",
    start_date="2020-01-01",  # Just get all ports
    end_date="2020-12-31"
)

if not ports_df.empty:
    print("\nSample ports:")
    print(ports_df[['portid', 'portname', 'country']].head(10))

    # Save port mapping for reference
    ports_df.to_csv('portwatch_port_ids.csv', index=False)
    print("\nPort IDs saved to 'portwatch_port_ids.csv'")

# 2. Now fetch daily RoRo data for FY24/25
print("\n" + "="*80)
print("FETCHING FY24/25 DAILY RORO DATA")
print("="*80)

# Use Daily_Trade_Data_WLD which has portcalls_roro field
roro_data = fetch_port_daily_data(
    service_name="Daily_Trade_Data_WLD",
    start_date="2024-04-01",
    end_date="2026-03-31"
)

if not roro_data.empty:
    print("\nData columns:", roro_data.columns.tolist())
    print("\nFirst few rows:")
    print(roro_data.head())

    # Save master file
    roro_data.to_csv('portwatch_fy24_25_roro.csv', index=False)
    print("\nData saved to 'portwatch_fy24_25_roro.csv'")

    # Summary statistics
    print("\n" + "="*80)
    print("FY24/25 DATA SUMMARY")
    print("="*80)

    # If we have portcalls_roro field
    if 'portcalls_roro' in roro_data.columns:
        summary = roro_data.groupby('portid').agg({
            'portcalls_roro': ['count', 'sum', 'mean'],
            'date': ['min', 'max']
        }).round(2)
        print("\nRoRo calls by port:")
        print(summary)

    # Check for Dover/Calais specifically
    if 'portid' in roro_data.columns:
        key_ports = roro_data[roro_data['portid'].isin(['PORT826', 'PORT132'])]  # Use actual IDs
        if not key_ports.empty:
            print("\nKey Channel ports data:")
            print(key_ports[['date', 'portid', 'portcalls_roro']].head())
```

```python
import requests
import pandas as pd
from datetime import datetime

def fetch_port_daily_data(port_ids=None, countries=None, start_date="2024-04-01", end_date="2026-03-31"):
    """
    Fetch daily port-level data from Daily_Ports_Data

    Parameters:
    - port_ids: List of port IDs (e.g., ['port1325', 'port1198']) or None for all ports
    - countries: List of country names or ISO3 codes to filter
    - start_date: YYYY-MM-DD
    - end_date: YYYY-MM-DD
    """

    base_url = "https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/Daily_Ports_Data/FeatureServer/0/query"

    # Build where clause - using year, month, day fields for efficient filtering
    start = datetime.strptime(start_date, '%Y-%m-%d')
    end = datetime.strptime(end_date, '%Y-%m-%d')

    # Create date range conditions using the separate fields
    date_conditions = []

    # For simplicity, we'll use the string date field which works with comparisons
    date_conditions = [f"date >= '{start_date}'", f"date <= '{end_date}'"]

    # Add port filters if specified
    if port_ids:
        port_list = "','".join(port_ids)
        date_conditions.append(f"portid IN ('{port_list}')")

    # Add country filters if specified
    if countries:
        country_list = "','".join(countries)
        date_conditions.append(f"country IN ('{country_list}') OR ISO3 IN ('{country_list}')")

    where_clause = " AND ".join(date_conditions)

    params = {
        "where": where_clause,
        "outFields": "date,year,month,day,portid,portname,country,ISO3,portcalls_roro,portcalls_container,portcalls_dry_bulk,portcalls_tanker",
        "returnGeometry": "false",
        "resultRecordCount": "5000",
        "maxRecordCountFactor": "5",
        "orderByFields": "portid ASC, date ASC",
        "f": "json"
    }

    print(f"\n📡 Fetching daily port data...")
    print(f"   Period: {start_date} to {end_date}")
    if port_ids:
        print(f"   Ports: {port_ids}")
    if countries:
        print(f"   Countries: {countries}")

    all_features = []
    offset = 0

    while True:
        params["resultOffset"] = str(offset)

        response = requests.get(base_url, params=params)

        if response.status_code != 200:
            print(f"❌ Error: {response.status_code}")
            break

        data = response.json()

        if 'error' in data:
            print(f"❌ API Error: {data['error'].get('message')}")
            break

        features = data.get('features', [])
        if not features:
            break

        all_features.extend(features)
        print(f"   Retrieved {len(all_features)} records so far...")

        if len(features) < 5000:  # Last page
            break

        offset += 5000

    if not all_features:
        print("❌ No data returned")
        return pd.DataFrame()

    # Convert to DataFrame
    records = []
    for f in all_features:
        attrs = f['attributes']
        # Convert date string to datetime
        if 'date' in attrs:
            attrs['date'] = pd.to_datetime(attrs['date'])
        records.append(attrs)

    df = pd.DataFrame(records)
    print(f"\n✅ Retrieved {len(df)} total records")

    return df

# Step 1: Get all ports to find UK and French port IDs
print("="*80)
print("STEP 1: Finding UK and French Port IDs")
print("="*80)

ports_url = "https://services9.arcgis.com/weJ1QsnbMYJlCHdG/arcgis/rest/services/PortWatch_ports_database/FeatureServer/0/query"
params = {
    "where": "country IN ('United Kingdom','France','Netherlands','Belgium')",
    "outFields": "portid,portname,country,ISO3,lat,lon",
    "returnGeometry": "false",
    "f": "json"
}

response = requests.get(ports_url, params=params)
ports_data = response.json()

europe_ports = []
for feat in ports_data.get('features', []):
    attrs = feat['attributes']
    europe_ports.append({
        'portid': attrs.get('portid'),
        'portname': attrs.get('portname'),
        'country': attrs.get('country'),
        'ISO3': attrs.get('ISO3')
    })

ports_df = pd.DataFrame(europe_ports)
print(f"\nFound {len(ports_df)} ports in UK, France, Netherlands, Belgium")
print("\nKey ports:")
key_port_names = ['Dover', 'Calais', 'Rotterdam', 'Zeebrugge', 'Antwerp', 'Felixstowe', 'Southampton']
key_ports = ports_df[ports_df['portname'].str.contains('|'.join(key_port_names), case=False, na=False)]
print(key_ports.to_string())

# Save port IDs for later
key_ports.to_csv('key_europe_ports.csv', index=False)
port_ids_list = key_ports['portid'].tolist()

# Step 2: Fetch FY24/25 daily data for these ports
print("\n" + "="*80)
print("STEP 2: Fetching FY24/25 Daily Port Data")
print("="*80)

daily_data = fetch_port_daily_data(
    port_ids=port_ids_list,
    start_date="2024-04-01",
    end_date="2026-03-31"
)

if not daily_data.empty:
    # Save master file
    daily_data.to_csv('fy24_25_port_daily_data.csv', index=False)
    print("\n💾 Saved to 'fy24_25_port_daily_data.csv'")

    # Summary statistics
    print("\n" + "="*80)
    print("📊 FY24/25 DATA SUMMARY")
    print("="*80)

    summary = daily_data.groupby(['portname', 'country']).agg({
        'portcalls_roro': ['sum', 'mean', 'count'],
        'date': ['min', 'max']
    }).round(2)

    print(summary)

    # Daily time series for key ports
    for port in daily_data['portname'].unique():
        port_df = daily_data[daily_data['portname'] == port].sort_values('date')
        print(f"\n📈 {port} daily RoRo calls:")
        print(port_df[['date', 'portcalls_roro']].head(10))
        print(f"... and {len(port_df)-10} more days")
```
