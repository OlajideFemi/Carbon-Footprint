```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wards = gpd.read_file("birmingham_wards.geojson")

wards = wards[[
    "WD22CD",   # ward code
    "WD22NM",   # ward name
    "geometry"
]].rename(columns={
    "WD22CD": "ward_code",
    "WD22NM": "ward_name"
})
```

```python
service = pd.read_csv("waste_service_metrics.csv")

service["miss_rate"] = (
    service["missed_collections"] /
    service["scheduled_collections"]
)
```

```python
gdf = wards.merge(
    service,
    on="ward_code",
    how="left"
)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

gdf.plot(
    column="miss_rate",
    ax=ax,
    legend=True,
    scheme="quantiles",
    edgecolor="black"
)

ax.set_title("Missed Waste Collection Rate by Ward â€“ Birmingham")
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

gdf.plot(
    column="backlog_tonnes",
    ax=ax,
    legend=True,
    scheme="quantiles",
    edgecolor="black"
)

ax.set_title("Estimated Waste Backlog (Tonnes) by Ward")
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 12))

gdf.plot(
    column="complaints",
    ax=ax,
    legend=True,
    scheme="quantiles",
    edgecolor="black"
)

ax.set_title("Resident Waste Complaints by Ward")
ax.axis("off")
plt.show()
```

```python
gdf["risk_index"] = (
    0.4 * gdf["miss_rate"] +
    0.4 * (gdf["backlog_tonnes"] / gdf["backlog_tonnes"].max()) +
    0.2 * (gdf["complaints"] / gdf["complaints"].max())
)
```
