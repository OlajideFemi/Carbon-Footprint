Simple Iterative Version

```python
def factorial_iterative(n):
  """"Calculate factorial using iteration"""
  if n < 0:
    return "Factorial is not defined for negative numbers"
  elif n == 0 or n == 1:
    return 1
  else:
    result = 1
    for i in range(2, n + 1):
      result *= i
    return result

print(factorial_iterative(5))
```

```python
try:
  num = int(input("Enter a non-negative integer: "))
  result = factorial_iterative(num)
  print(f"The factorial of {num} is {result}")
except ValueError:
  print("Invalid input. Please enter a non-negative integer.")
```

Recursive Version

```python
def factorial_recursive(n):
    """Calculate factorial using recursion"""
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

# Get input from user
try:
    num = int(input("Enter a number to find its factorial: "))
    result = factorial_recursive(num)
    print(f"The factorial of {num} is: {result}")
except ValueError:
    print("Please enter a valid integer")
```

One-Liner Using Python's Built-in Function

```python
import math

# The simplest way
num = int(input("Enter a number: "))
print(f"The factorial of {num} is: {math.factorial(num)}")
```

```python
def factorial_descending(n):
    """Calculate and display factorial in descending order"""
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        print("0! = 1")
        return 1

    result = 1
    # Build the expression string in descending order
    expression = " × ".join(str(i) for i in range(n, 0, -1))

    # Calculate the factorial
    for i in range(n, 0, -1):
        result *= i

    print(f"{n}! = {expression} = {result}")
    return result

# Test the function
num = int(input("Enter a number: "))
factorial_descending(num)
```

```python
def factorial_step_by_step_descending(n):
    """Show each step of factorial calculation in descending order"""
    if n < 0:
        print("Factorial is not defined for negative numbers")
        return None

    print(f"\nCalculating {n}! in descending order:")
    print("-" * 40)

    result = 1
    for i in range(n, 0, -1):
        result *= i
        if i == n:
            print(f"Step {n}: {i} = {result}")
        else:
            print(f"Step {i}: {result} × {i-1 if i-1 > 0 else ''} = {result * (i-1) if i-1 > 0 else result}")
            # Alternative simpler display:
            # print(f"Step: {i} × ... = {result}")

    print("-" * 40)
    print(f"Final result: {n}! = {result}")
    return result

# Test
num = int(input("Enter a number: "))
factorial_step_by_step_descending(num)
```

```python
def factorial_with_descending_display(n, show_details=True):
    """Display factorial with descending order options"""

    if n < 0:
        print("Error: Factorial not defined for negative numbers")
        return None

    print(f"\n{'='*50}")
    print(f"FACTORIAL OF {n} (DESCENDING ORDER)")
    print(f"{'='*50}")

    # Calculate factorial
    result = 1
    numbers = list(range(n, 0, -1))

    # Option 1: Show complete expression
    expression = " × ".join(map(str, numbers))
    print(f"\nComplete expression:")
    print(f"{n}! = {expression}")

    # Option 2: Show with calculation steps
    if show_details:
        print(f"\nStep-by-step calculation:")
        temp = 1
        for i, num in enumerate(numbers, 1):
            temp *= num
            remaining = numbers[i:] if i < len(numbers) else []
            if remaining:
                remaining_expr = " × ".join(map(str, remaining))
                print(f"Step {i}: {temp} × {remaining_expr} = {temp * (temp if i==1 else 1)}")
            else:
                print(f"Step {i}: {temp} (final)")

    # Option 3: Show final result with digit count
    print(f"\n{'─'*50}")
    print(f"Final Result: {n}! = {result}")
    print(f"Number of digits: {len(str(result))}")
    print(f"{'='*50}")

    return result

# Test with user input
try:
    num = int(input("Enter a non-negative integer: "))
    factorial_with_descending_display(num, show_details=True)
except ValueError:
    print("Please enter a valid integer")
```

```python
def factorial_recursive_descending(n, original_n=None):
    """Recursive factorial with descending display"""
    if original_n is None:
        original_n = n

    if n < 0:
        return "Error: Negative input"

    if n == 0 or n == 1:
        if n == 0:
            print(f"0! = 1")
        return 1

    # Display in descending order
    print(f"{n} × ", end="")
    result = n * factorial_recursive_descending(n - 1, original_n)

    # Print final result at the end
    if n == original_n:
        print(f"1 = {result}")

    return result

# Test
num = int(input("Enter a number: "))
print(f"\n{num}! = ", end="")
factorial_recursive_descending(num)
```

```python
def factorial_table_descending(n):
    """Display factorial in a table format with descending order"""

    if n < 0:
        print("Invalid input")
        return

    print(f"\n{'='*60}")
    print(f"FACTORIAL TABLE FOR {n} (DESCENDING)")
    print(f"{'='*60}")

    # Calculate and store intermediate results
    results = []
    current = 1
    for i in range(1, n + 1):
        current *= i
        results.append(current)

    # Display in descending order
    print(f"{'Step':<6} {'Expression':<30} {'Result':<15}")
    print("-" * 60)

    for i in range(n, 0, -1):
        # Build expression
        if i == n:
            expr = f"{n}"
        else:
            expr = f"{n} × ... × {i}"

        # Get result
        res = results[i-1] if i-1 < len(results) else 1

        print(f"{n-i+1:<6} {expr:<30} {res:<15,}")

    print("-" * 60)
    print(f"Final: {n}! = {results[-1]:,}")
    print(f"{'='*60}")

# Test
num = int(input("Enter a number: "))
factorial_table_descending(num)
```

```python
import time

def factorial_with_animation(n):
    """Animated display of factorial in descending order"""

    if n < 0:
        print("Invalid input")
        return

    result = 1
    print(f"\nCalculating {n}! in descending order:")
    print("-" * 40)

    for i in range(n, 0, -1):
        result *= i
        # Display current multiplication
        factors = [str(x) for x in range(n, i-1, -1)]
        expr = " × ".join(factors)
        print(f"\r{expr} = {result}", end="")
        time.sleep(0.5)  # Pause for animation effect

    print(f"\n\nFinal: {n}! = {result}")
    print("-" * 40)

# Test
num = int(input("Enter a number: "))
factorial_with_animation(num)
```
