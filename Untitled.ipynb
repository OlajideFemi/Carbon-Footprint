<p>Deriving PCA </p>

<p>$x_{1} = Y_{W}$</p>
<p>Generally for and random variable the rule is</p>
<p>$var(a) = E(a^{2}) - E(a)^{2}$ </p>
<p>so lets continue with our example and find the $var(x_{1})$, but because its a vector, our notation will be slightly different.</p>

<p>$var(x_{1} ) = E(x_{1}^{T} x_{1} ) - E(x_{1})^{T}E(x_{1})$ </p>
<p>$ = E(Y_{W}^{T} Y_{W} ) - E(Y_{W})^{T}E(Y_{W})$ </p>
<p>$ = E(W^{T}Y^{T} Y_{W} ) - E(W^{T}Y^{T}) E(Y_{W})$ </p>
<p>another rule to remember, let $\alpha$ be a scalar, and let $x$ be a random variable, then </p>
<p>E($\alpha x$) = $\alpha E(x)$ </p>
<p>then</p>
<p>$var(x_{1} ) = W^{T}E(Y^{T}Y )W - W^{T}E(Y^{T})E(Y)W$ </p>
<p>now we know that $ E(Y) = 0 $ so then</p>

<p>$var(x_{1} ) = W^{T}E(Y^{T}Y )W$ </p>
<p>$ = W^{T}\sum W$ </p>
<p>and we want to maximise the $var(x_{1} )$, and we can only change $W$ because $\sum$ is fixed. Now the problem is that if we simply 'maximize' then by setting $W=\infty$ would solve the problem. But this makes no sense, so we add a constraint. We make use out of the Lagrangian function  </p>
<p>$\mathcal{L}$ $(W,\lambda)$ = $W^{T} \sum W - \lambda (W^{T}W - 1)$</p>
<p>and now we maximise $\mathcal{L}$ $(W,\lambda)$ with respect to $W$</p>
$\frac{d \mathcal{L} (W,\lambda)}{dW}$ = $2 \sum W - 2 \lambda W$
<p> and we set this to zero. We find </p>
<p> $2 \sum W - 2 \lambda W$ = 0</p>
<p>$2 \sum W = 2 \lambda W $</p>
<p>$ \sum W =  \lambda W $</p>
<p>for those who remember linear algebra would relaise that $W$ is the largest (with respect to the eigenvalue) eigenvector of $\sum$.  So going back to our original definition we re-write $x_{1} = Y_{W}$ as  </p>
$$x_{1} = Y_{W_{1}}$$


<p>But do we really just want 1 dimensional projection?</p>
<p>Finding the second dimension of the projection.</p>

$$
X = [x_{1},x_{2}]
$$


$$X = \begin{bmatrix} x_{11} & x_{12} \\ x_{21} & x_{22} \\ \vdots & \vdots \\ x_{N1} & x_{N2}
\end{bmatrix}$$

then

$$
x_{2} = Y_{W_{2}}
$$
$$
var(x_{2}) = E(x^{T}_{2} x_{2}) - E(x_{2})^{T} E(x_{2})
$$

$$
= E((Y_{W_{2}})^{T} Y_{W_{2}}) - E
(Y_{W_{2}})^{T} E(Y_{W_{2}})
$$

$$
= E((_{W_{2}^{T}}Y^{T}Y_{W_{2}}) - E
(_{W_{2}^{T}}Y^{T}) E(Y_{W_{2}})
$$


So we want to maximize this too, and we propose to use the same method as before, except we also add the constraint $w_{1}w_{2}^{T} = 0.$ So the largrangian function used

$$
\mathcal{L}(w_{2},\lambda_{2}) = w_{2}^{T}\sum w_{2} - \lambda_{2} (w_{2}^{T}w_{2} - 1)
$$

and we set this to zero. We find
$$
2 \sum w_{2} - 2 \lambda_{2}w_{2} = 0
$$

$$
2 \sum w_{2} =  2 \lambda_{2}w_{2}
$$

$$
\sum w_{2} =  \lambda_{2}w_{2}
$$


from linear algebra we would realize that $w_{2}$ is the second largest (with respect to the eigenvalue) eigenvector of $\sum$, this is because $w_{1}$ is the largest.
And therefore we can write
$$
X = YW
$$
where
$$
W = [w_{1}, w_{2}, \dots , w_{D} ]
$$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

def plot_svc_decision_function(clf, ax=None):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = clf.decision_function(xy).reshape(X.shape)

    # Plot decision boundary and margins
    ax.contour(X, Y, P, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
    ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

    # Add axis labels and title
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_title(f"SVM decision boundary (C={clf.C})")


# Load the dataset
df = pd.read_csv('training.csv')

# Separate the features and target variable
X = df.iloc[:, :2].values
y = df.iloc[:, 2].values

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Define the SVM model
C = 1
svm = SVC(kernel='linear', C=C)

# Train the SVM model
svm.fit(X, y)

# Plot the decision boundary in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create grid to evaluate model
x = np.linspace(X[:, 0].min(), X[:, 0].max(), 30)
y = np.linspace(X[:, 1].min(), X[:, 1].max(), 30)
X1, X2 = np.meshgrid(x, y)
Z = np.zeros_like(X1)

# Compute decision function for each point in the grid
for i in range(X1.shape[0]):
    for j in range(X1.shape[1]):
        Z[i, j] = svm.predict([[X1[i, j], X2[i, j]]])

# Plot decision surface
surf = ax.plot_surface(X1, X2, Z, cmap=plt.cm.coolwarm, alpha=0.5)

# Plot support vectors
ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], svm.predict(svm.support_vectors_), s=100, linewidth=1, facecolors='none', edgecolors='k')

# Add decision boundary in 2D
plot_svc_decision_function(svm, ax=ax)

# Set the plot title and axis labels
ax.set_title(f"SVM decision boundary (C={C})")
ax.set_xlabel('Feature 1 (standardized)')
ax.set_ylabel('Feature 2 (standardized)')
ax.set_zlabel('Target variable')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import classification_report

# load training data
train_data = pd.read_csv('training.csv')
X_train = train_data.iloc[:, :2].values
y_train = train_data.iloc[:, 2].values

# train SVM with different values of C
C_values = list(range(1, 11))
models = []
for C in C_values:
    clf = svm.SVC(C=C, kernel='linear')
    clf.fit(X_train, y_train)
    models.append(clf)

# load validation data
val_data = pd.read_csv('validation-.csv')
X_val = val_data.iloc[:, :2].values
y_val = val_data.iloc[:, 2].values

# divide validation set into 10 subsets
subset_size = len(X_val) // 10
X_subsets = [X_val[i:i+subset_size] for i in range(0, len(X_val), subset_size)]
y_subsets = [y_val[i:i+subset_size] for i in range(0, len(y_val), subset_size)]

# evaluate models on each subset
errors = []
for clf in models:
    subset_errors = []
    for X, y in zip(X_subsets, y_subsets):
        y_pred = clf.predict(X)
        subset_errors.append(1 - clf.score(X, y))
    errors.append(subset_errors)

# calculate mean and standard deviation of errors
mean_errors = np.mean(errors, axis=1)
std_errors = np.std(errors, axis=1)

# print mean and standard deviation for each C
for C, mean_err, std_err in zip(C_values, mean_errors, std_errors):
    print(f'C={C}, mean error={mean_err}, std error={std_err}')

# plot mean and standard deviation of errors
import matplotlib.pyplot as plt
plt.errorbar(C_values, mean_errors, yerr=std_errors, fmt='o')
plt.xlabel('C values')
plt.ylabel('Mean error with standard deviation')
plt.show()

plt.plot(C_values, mean_errors, 'o-')
plt.xlabel('C values')
plt.ylabel('Mean error')
plt.show()

plt.plot(C_values, std_errors, 'o-')
plt.xlabel('C values')
plt.ylabel('Standard deviation of error')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV

# Load data
train_data = pd.read_csv('training.csv')
X_train = train_data.iloc[:, :2].values
y_train = train_data.iloc[:, 2].values

# Define parameter grid for grid search
param_grid = {'C': list(range(1, 11)),
              'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
              'degree': list(range(1, 6))}

# Create SVM classifier object
svm_clf = svm.SVC()

# Create GridSearchCV object with SVM classifier and parameter grid
grid_search = GridSearchCV(svm_clf, param_grid, cv=5)

# Train the GridSearchCV object on the training data
grid_search.fit(X_train, y_train)

# Print best parameters and mean accuracy score
print(f"Best parameters: {grid_search.best_params_}")
print(f"Mean accuracy score: {grid_search.best_score_:.3f}")
```
